linear regression with one variable

```python
data = np.loadtxt('ex1data1.txt', delimiter=',')

#create matrix X with first column values 1 and second column values x1:
X = np.c_[np.ones(data.shape[0]),data[:,0]]
# la
y = np.c_[data[:,1]]

plt.scatter(X[:,1], y, s=30, c='r', marker='x', linewidths=1)
plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
```

```python
def compute_cost(X, y, theta=[[0],[0]]):
    theta=np.array(theta)
    m=y.size  # size will return row * column in this case m*1
    J=0.0
    #use vectorization compute the XTheta.T*XTheta = (hx-y)**2
    error = X.dot(theta)-y
    J=1.0/(2*m)*((error.T).dot(error))
    return J
```

```python
compute_cost(X, y)
```

```python
def gradient_descent(X, y, theta=[[0],[0]], alpha=0.01, num_iters=1500):
    m=y.size
    J_history = np.zeros(num_iters)
    theta = np.array(theta)
    for i in np.arange(num_iters):
        #theta =theta-alpha*1/m*(Xtheta.T*X).T
        theta = theta-alpha*(1.0/m)*(((X.dot(theta)-y).T).dot(X)).T
        J_history[i] = compute_cost(X,y,theta)
    return theta, J_history
```

```python
# theta for minimized cost J
theta , Cost_J = gradient_descent(X, y)
print theta
print('theta: ',theta.ravel()) # 
itr = np.arange(1500)
plt.plot(itr, Cost_J)
plt.ylabel('Cost J')
plt.xlabel('Iterations');
```

```python
xx = np.arange(5,23)
yy = theta[0]+theta[1]*xx

# Plot gradient descent
plt.scatter(X[:,1], y, s=30, c='r', marker='x', linewidths=1)
plt.plot(xx,yy, label='Linear regression (Gradient descent)')

#compare above line with sklearn package
regr = LinearRegression()
regr.fit(np.c_[X[:, 1]], y.ravel())
plt.plot(xx, regr.intercept_+regr.coef_*xx)
```

```python
# Predict profit for a city with population of 35000 and 70000
print(theta.T.dot([1, 3.5])*10000)
print(theta.T.dot([1, 7])*10000)
#predict with sklearn 
print(regr.intercept_+regr.coef_*3.5)
print(regr.intercept_+regr.coef_*7)
```
